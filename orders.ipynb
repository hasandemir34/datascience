```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("datasetler/orders.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["year"].value_counts()
```

```python
df.drop(["order_id","customer_name","product_id","product_name"],axis=1,inplace=True)
```

```python
df["order_date"].unique()
```

```python
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)
#object olan tarihleri tarih formatına getirdim.
```

```python
df["time"]=df["ship_date"]-df["order_date"]
```

```python
df.drop(["order_date","ship_date"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.time.value_counts()
```

```python
df["time"] = df["time"].dt.days
```

```python
df.info()
```

```python
df.ship_mode.value_counts()
```

```python
ship_mapping = {
    'Standard Class': 0,
    'Second Class': 1,
    'First Class': 2,
    'Same Day': 3
}
```

```python
df["ship_mode"]=df["ship_mode"].map(ship_mapping)
```

```python
df.segment.value_counts()
```

```python
df = pd.get_dummies(df, columns=['segment'], prefix='seg')
```

```python
df.head()
```

```python
df.order_priority.value_counts()
```

```python
onem={
    "Low":0,
    "Medium":1,
    "High":2,
    "Critical":3
}
```

```python
df["order_priority"]=df["order_priority"].map(onem)
```

```python
df.head()
```

```python
df.market.value_counts()
```

```python
df.region.value_counts()
```

```python
df.sub_category.value_counts()
```

```python
df.country.value_counts()
```

```python
#market ve region bölgeleri için one hot yapıcam
df = pd.get_dummies(df, columns=['market', 'region'], prefix=['mkt', 'reg'])
```

```python
df.info()
```

```python
#countryleri de ülkenin ortalama değerine bölerek hesaplama yapıcam. yani türkiyenin ortalama değeri neyse artık
#türkiye yerine mesela 49 yazacak

country_means = df.groupby('country')['profit'].mean()
df['country'] = df['country'].map(country_means)
```
